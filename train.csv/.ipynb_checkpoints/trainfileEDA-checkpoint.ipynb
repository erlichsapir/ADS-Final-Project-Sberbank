### Missing values

```python
basic_house_missing = basic_house_data.isnull().sum()/len(basic_house_data)
basic_house_missing = basic_house_missing.drop(basic_house_missing[basic_house_missing == 0].index).sort_values(ascending = False).reset_index()
basic_house_missing.columns = ['column name','missing percentage']
print(basic_house_missing)
```

### Outliers/errors detection

```python
groupy_count = basic_house_data.groupby('build_year')['id'].aggregate('count').reset_index()
groupy_count.columns=['build_year','count']
print(groupy_count.head(10))
print(groupy_count.tail(10))
```

Build years below 1800 and above 2018 are probably mistakes (or maybe NaN, for the values 0 and 1)

```python
groupy_count = basic_house_data.groupby('state')['id'].aggregate('count').reset_index()
groupy_count.columns=['state','count']
print(groupy_count.head())
```

33 is probably ment to be 3

```python
groupy_count = basic_house_data.groupby('num_room')['id'].aggregate('count').reset_index()
groupy_count.columns=['num_room','count']
print(groupy_count)
```

```python
groupy_count = basic_house_data.groupby('product_type')['id'].aggregate('count').reset_index()
groupy_count.columns=['product_type','count']
print(groupy_count)
```
